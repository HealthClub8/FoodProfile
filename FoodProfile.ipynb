```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you already have the `total_and_rda` DataFrame with 'Total' and 'RDA' columns

# Step 1: Calculate the percentage of total vs RDA for each nutrient
total_and_rda['Percentage'] = (total_and_rda['Total'] / total_and_rda['RDA']) * 100

# Step 2: Cap the percentage at 200% to avoid values exceeding the chart's range
total_and_rda['Percentage'] = total_and_rda['Percentage'].apply(lambda x: min(x, 1000))

# Step 3: Sort the DataFrame by Percentage in descending order
total_and_rda_sorted = total_and_rda.sort_values(by='Percentage', ascending=True)

# Step 4: Plotting the sorted horizontal bar chart
plt.figure(figsize=(10, 8))

# Create horizontal bars for the percentage of total vs. RDA using gray color
plt.barh(total_and_rda_sorted.index, total_and_rda_sorted['Percentage'], color='grey', label='Total % of RDA')

# Create a black bar for the RDA value (which represents 100%)
for i, rda in enumerate(total_and_rda_sorted['RDA']):
    plt.barh(i, 100, color='red', height=0.5)  # Full bar for RDA (100%)
    
    # Create a yellow bar for the total percentage (relative to the RDA)
    plt.barh(i, total_and_rda_sorted['Percentage'].iloc[i], color='grey', height=0.5)

# Add labels and title
plt.xlabel('Percentage of RDA (%)')
plt.title('Total as Percentage of RDA for Each Vitamin/Nutrient (Ordered by Highest)')

# Display the plot
plt.tight_layout()
plt.show()
```
